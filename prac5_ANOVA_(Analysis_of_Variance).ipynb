**ANOVA (F-TEST)**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv("/content/Covid Data.csv")
```

```python
# Keep only PATIENT_TYPE and AGE
df_anova = df[['PATIENT_TYPE', 'AGE']]

# Show first rows
print(df_anova.head())
```

```python
# Identify unique group names
grps = pd.unique(df_anova.PATIENT_TYPE.values)
```

```python
# Create dictionary of group: values
d_data = {grp: df_anova['AGE'][df_anova.PATIENT_TYPE == grp].dropna()
          for grp in grps}
```

```python
# Ensure at least 2 groups (ANOVA needs 2+ groups)
if len(d_data) >= 2:
    # Unpack the dictionary values (the age groups) into f_oneway
    F, p = stats.f_oneway(*d_data.values())
else:
    raise ValueError("Your dataset does not have at least 2 groups for ANOVA. One-way ANOVA requires at least two groups for comparison.")

print("p-value for significance is: ", p)

if p < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

**Two Way F-test**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv("/content/Covid Data.csv")
```

```python
# Keep only needed columns
df_anova2 = df[['AGE', 'PATIENT_TYPE', 'SEX']].dropna()
# Two-way ANOVA: AGE explained by PATIENT_TYPE and SEX
model = ols('AGE ~ C(PATIENT_TYPE) * C(SEX)', data=df_anova2).fit()
# Print overall model F-statistic and p-value
print(f"Overall model F({model.df_model:.0f},{model.df_resid:.0f}) = "
      f"{model.fvalue:.3f}, p = {model.f_pvalue:.4f}")
```

```python
# ANOVA table (Type II)
res = sm.stats.anova_lm(model, typ=2)
print(res)
```
