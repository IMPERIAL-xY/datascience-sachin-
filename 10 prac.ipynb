```python
# ================================
# ðŸ“Œ DATA VISUALIZATION PRACTICAL
# ================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To display plots inside notebook
%matplotlib inline

# Load dataset
df = pd.read_csv("Heart.csv")

# Preview dataset
df.head()
```

```python
df.describe()
```

```python
plt.plot(df["MaxHR"].head(200))
plt.title("Line Plot â€“ Maximum Heart Rate Trend")
plt.xlabel("Patient Index")
plt.ylabel("Maximum Heart Rate")
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=df["AHD"])
plt.title("Bar Plot â€“ Heart Disease Distribution")
plt.xlabel("Heart Disease (AHD)")
plt.ylabel("Count")
# plt.show()
```

```python
# Create the histogram and density plot for 'Age'
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Histogram + Density â€“ Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count / Density")
# Removed plt.savefig() and plt.close() to display in output
plt.show()
```

```python
# Create the KDE plot for 'Chol'
plt.figure(figsize=(8,5))
sns.kdeplot(df["Chol"], fill=True)
plt.title("Distribution Plot â€“ Serum Cholesterol")
plt.xlabel("Serum Cholesterol (mg/dl)")
plt.ylabel("Density")
# Removed plt.savefig() and plt.close() as requested
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["MaxHR"], df["Age"], alpha=0.5)
# Update the title and labels
plt.title("Scatter Plot â€“ Maximum Heart Rate vs Age")
plt.xlabel("Maximum Heart Rate (MaxHR)")
plt.ylabel("Age")
# Removed plt.savefig() as requested
plt.show()
```

```python
plt.figure(figsize=(6,6))
# Use 'Sex' column for the pie chart
df["Sex"].value_counts().plot.pie(
    autopct="%1.1f%%",
    labels=["Male (1)", "Female (0)"], # Add explicit labels for clarity
    startangle=90
)
# Update the title
plt.title("Pie Chart â€“ Sex Distribution")
plt.ylabel("")
# Removed plt.savefig() as requested
plt.show()
```

```python
counts = df["AHD"].value_counts()
labels = counts.index.tolist()
# Explode the 'Yes' slice (assuming 'No' is the majority and comes first)
explode_list = [0, 0.1] # [No, Yes]

plt.figure(figsize=(6,6))
plt.pie(
    counts,
    labels=labels,
    autopct="%1.1f%%",
    explode=explode_list,
    startangle=90
)
plt.title("Exploded Pie Chart â€“ Heart Disease Diagnosis (AHD)")
plt.ylabel("")
# plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x=df["Age"])
plt.title("Box Plot â€“ Age Distribution")
plt.xlabel("Age")
# Removed plt.savefig() as requested
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.violinplot(x=df["Age"])
plt.title("Violin Plot â€“ Age Distribution")
plt.xlabel("Age")
# Removed plt.savefig() as requested
plt.show()
```

```python


# Create the combined box and violin plot using the correct categorical variable "AHD"
plt.figure(figsize=(8,6))

# 1. Violin Plot
sns.violinplot(
    x=df["AHD"], # Corrected column
    y=df["Age"],
    inner=None,
    color="lightblue"
)

# 2. Box Plot (overlaid)
sns.boxplot(
    x=df["AHD"], # Corrected column
    y=df["Age"],
    width=0.2,
    color="white" # Use a distinct color for the box plot for visibility
)

# Update the title and labels
plt.title("Combined Box & Violin â€“ Age vs Heart Disease (AHD)")
plt.xlabel("Heart Disease Diagnosis (AHD)")
plt.ylabel("Age")

# Save the plot
plt.savefig("age_vs_ahd_combined_plot.png")
plt.close()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True, bins=30)
plt.title("Histogram + Density â€“ Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count / Density")
# Removed plt.savefig() and using plt.show() as requested
plt.show()
```

```python
# Fabricate the time series data for the first 200 rows
df_ts = df.head(200).copy()
df_ts["JoinDate"] = pd.date_range(start="2020-01-01", periods=200, freq="D")

# Create the time series plot using the correct column 'Chol'
plt.figure(figsize=(10,5))
plt.plot(df_ts["JoinDate"], df_ts["Chol"]) # Corrected column
plt.title("Time Series â€“ Serum Cholesterol Over Time") # Corrected title
plt.xlabel("Join Date")
plt.ylabel("Serum Cholesterol (mg/dl)") # Corrected y-label
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
